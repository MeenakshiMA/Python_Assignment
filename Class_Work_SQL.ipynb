```python
import mysql.connector as conn
```

```python
mydb = conn.connect(host = 'localhost', user = 'root', passwd= 'Pas$w0rd@123')
```

```python
cursor = mydb.cursor()
```

```python
#create a database called cardataset
cursor.execute('create database cardataset')
```

```python
#create a table called as car with column name given in dataset description
cursor.execute("create table cardataset.car (buying VARCHAR(30), maint VARCHAR(30), doors VARCHAR(10), persons VARCHAR(10), lug_boot VARCHAR(30), safety VARCHAR(30), none VARCHAR(30))")
```

```python
# dump all the data into car table
import csv
with open('car.data', 'r') as f :
    car_data = csv.reader(f, delimiter = '\n')
    for i in car_data:
        s = i[0].split(",")
        query = "INSERT INTO cardataset.car VALUES('{}','{}','{}','{}','{}','{}','{}')" . format(str(s[0]),str(s[1]),str(s[2]),str(s[3]),str(s[4]),str(s[5]),str(s[6]))
        cursor.execute(query)
mydb.commit()
```
